```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("diabetes.csv", delimiter=',')

dataset= np.genfromtxt("diabetes.csv",delimiter=',',skip_header=True)
np.set_printoptions(formatter={'float':'{: 0.1f}'.format})
print(dataset[:5])
print(data.shape)
```

```python
X =dataset[:,:-1]
Y=  dataset[:,-1]
print(X.shape,Y.shape)
```

* IMPORT Sequential, Dense from keras

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#! Create model
```

```python
model = Sequential()
model.add(Dense(1,input_dim = len(X[0,: ]),activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```python
history= model.fit(x=X,y=Y,epochs=1000,batch_size=32,verbose=0)
```

```python
predictions = model.predict(X)
```
